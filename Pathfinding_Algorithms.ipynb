```python
def gen_map(x, y, variance, visualise = True, zeroes = False):
    '''
    Returns a random sqaure matrix of size x as a map for pathfinding.
    Values are assigned randomly between 0 and variance.
    0s are walls, impassible. 
    Also returns a boolean map of 0s
    ''' 
    initial = np.random.randint(0, variance + 1, (x, y)) 
    if visualise == True:
        plt.imshow(initial)
        plt.show()
    
    if zeroes == True:
        visited = np.zeros((x,x), dtype = int)
        return(initial, visited)
    else:
        return(initial)
```

```python
#Node class for dijkstra and greedy first
class Node:
    '''
    A node class. Contains information on distance and position.
    '''
    def __init__(self, r:int, c:int, dist:int):
        self.loc = (r,c)
        self.dist = dist
        
    def __str__(self):
        return(f'{self.loc}, {self.dist}')
    
    def __repr__(self):
        return(f'<{self.loc} {self.dist}>')
    
    # These two allow for min and max functions to work.
    def __lt__(self, other):
        return self.dist < other.dist
    
    def __gt__(self, other):
        return self.dist > other.dist
```

```python
#Node class for A* algorithm
class StarNode:
    '''
    A Node class for A*. Contains information on distance and position.
    If viewed, example is <(0,0) d5+e2=7>, where (x,y) are co-ordiantes,
    d5 = distance of node or distance from start node (g),
    e2 = estimate of distance to end node (h),
    7 = total distance of d and e (f = g + h)
    '''
    def __init__(self, r:int, c:int, dist:int, est:int = 0):
        self.loc = (r,c)
        self.dist = dist
        self.est = est
        self.total = dist + est
        
    def __str__(self):
        return(f'{self.loc}, d{self.dist}+e{self.est}={self.total}')
    
    def __repr__(self):
        return(f'<{self.loc} d{self.dist}+e{self.est}={self.total}>')
    
    # These two allow for min and max functions to work.
    def __lt__(self, other):
        return self.total < other.total
    
    def __gt__(self, other):
        return self.total > other.total
    
    # Allows modification of the estimate along with updating the total.
    def new_est(self, x):
        self.est = x
        self.total = self.est + self.dist
```

```python
#Code for Dijkstra algorithm, uses Node class.

def neighbours_dijkstra(n:Node, terrain, visited, check:dict = None): 
    '''
    Finds all valid neighbour nodes next to target node(n).
    Takes in a map (terrain) and corresponding bool map (visited).
    Only works within the search function.
    Returns a list of node classes that are neighbour nodes to the input node.
    '''
    # Create a new node for each direction
    north = Node(n.loc[0] - 1, n.loc[1], n.dist)
    south = Node(n.loc[0] + 1, n.loc[1], n.dist)
    east = Node(n.loc[0], n.loc[1] + 1, n.dist)
    west = Node(n.loc[0], n.loc[1] - 1, n.dist)
    neighbour_list = []
    shape = terrain.shape
    length, width = shape[0], shape[1]
    
    for i in [north, south, west, east]:
        direction = (i.loc[0], i.loc[1])
        r,c = direction[0], direction[1]
        # Checking that the new nodes are not out of bounds
        if r != -1 and c != -1 and r != length and c != width:
            # Checking that the new nodes are not walls or previously visited.
            if terrain[direction] != 0 and visited[direction] == 0:
                temp = Node(i.loc[0], i.loc[1],
                            (i.dist + terrain[i.loc]))
                neighbour_list.append(temp)
            # This allows for nodes to be overriden at a location, if they
            # have a smaller distance than the existing node at that location.
            elif i.loc in check:
                x = check[i.loc]
                if (int(i.dist) + terrain[i.loc]) < int(x.dist):
                    if terrain[direction] != 0 and visited[direction] == 1:
                        temp = Node(i.loc[0], i.loc[1],
                                    (i.dist + terrain[i.loc]))
                        neighbour_list.append(temp)
    return(neighbour_list)


def dijkstra(startnode:Node, terrain, endpoint = None, everything = False,
             return_visited = False):
    '''
    startnode:node -> The node at which pathfinding begins from.
    terrain -> A numpy array representing the map through which pathfinding
               will occur.
    endpoint -> A co-ordiante (x,y) on the graph to which the most optimum path
                will be found for.
    everything -> Will go through every possible path. Default is False.
                  Realistically should never be set to True.
    return_visited -> will return count of how many nodes visited.
    Function returns; tuple(A list of co-ordianates representing the path,
                            distance of the path)

    '''
    shape = terrain.shape
    visited = np.zeros((shape), dtype = int) 
    pathlist = [[startnode]]
    nodedict = {startnode.loc:startnode}
    start = True
    finished = False
    while finished == False:

        if start == True:
            queue = [startnode]
            visited[startnode.loc] = 1
            start = False
        # Break out of loop when reaching destination
        if everything == False and queue[0].loc == endpoint:
            break
        # A queue system to determine which node's neighbours get checked next.
        temp = neighbours_dijkstra(queue[0], terrain, visited, nodedict)
        for i in temp:
            visited[i.loc] = 1
            queue.append(i)
            # Dictionary to record visited nodes and which current node occupies
            # the location.
            if i in nodedict.keys():
                nodedict.pop(i.loc)
            nodedict.update({i.loc:i})
        
        #This loop creates a list of every path taken.
        for path in pathlist:
            for i in range(0, len(temp)):
                x = path[:]
                if queue[0] == x[-1]:
                    x.append(temp[i])
                    pathlist.append(x)
        queue.pop(0)
        
        if everything == False:
            queue.sort()
        # A patchwork solution of removing nodes from the queue, if a node
        # with the same location but a smaller distance exists in the queue.
        for node in queue:
            for node2 in queue:
                if node.loc == node2.loc and node.dist < node2.dist:
                    queue.remove(node2)
                else:
                    continue
        # If there are no more possible paths, terminate.
        if len(queue) == 0:
            finished = True

    # This finds the path that has the endpoint and has the lowest distance.
    final_list = []
    first = True
    for path in pathlist:
        if path[-1].loc == endpoint:
            if first == True:
                final_list = path[:]
            elif path[-1].dist < final_list[-1].dist:    
                final_list = path[:]
    
    # Reformates paths into pure tuple co-ordiantes and also returns distance
    # seperately.
    final_path = []
    for node in final_list:
        final_path.append(node.loc)        
    if len(final_list) != 0:
        final_dist = final_list[-1].dist
    else:
        final_dist = 'Null'
    if return_visited == False:
        return(final_path, final_dist)
    if return_visited == True:
        unique, counts = np.unique(visited, return_counts=True)
        x = dict(zip(unique, counts))
        total = x[1]
        return(final_path, final_dist, total)
        
```

```python
#Code for greedy algorithm, uses Node class

def neighbours_greedy(n:Node, terrain, visited,
                      random = False, endpoint = None): 
    '''
    Finds all valid neighbour nodes next to target node(n).
    Takes in a map (terrain) and corresponding bool map (visited).
    Only works within the search function.
    
    Random -> If set to True, tie breaks are random, otherwise tie breaks
    favour nodes with closest manhattan distance, in such cases with more than
    one occurrence, tie breaks will still be random.
    
    Returns the node that is most optimum.
    '''
    # Create a new node for each direction
    north = Node(n.loc[0] - 1, n.loc[1], n.dist)
    south = Node(n.loc[0] + 1, n.loc[1], n.dist)
    east = Node(n.loc[0], n.loc[1] + 1, n.dist)
    west = Node(n.loc[0], n.loc[1] - 1, n.dist)
    neighbour_list = []
    temp_list = []
    shape = terrain.shape
    length, width = shape[0], shape[1]
    
    for i in [north, south, west, east]:
        direction = (i.loc[0], i.loc[1])
        r,c = direction[0], direction[1]
        # Checking that the new nodes are not out of bounds.
        if r != -1 and c != -1 and r != length and c != width:
            # Checking that the new nodes are not walls or previously visited.
            if terrain[direction] != 0 and visited[direction] == 0:
                temp = Node(i.loc[0], i.loc[1],
                            (i.dist + terrain[i.loc]))
                neighbour_list.append(temp)
    
    # Checks if any neighbours exist.
    if len(neighbour_list) == 0:
        return(None)
    
    # This section chooses the node with the smallest distance and any
    # other points with an equal distance.
    smallest_node = min(neighbour_list)
    neighbour_list.remove(smallest_node)
    temp_list.append(smallest_node)
    for neighbour_node in neighbour_list:
        if neighbour_node.dist == temp_list[0].dist:
            temp_list.append(neighbour_node)
            
    # Tie break if endpoint is known and random is false. Prefers nodes that
    # are closest to final node by absolute distance, if another tie, random.
    if len(temp_list) > 1 and random == False and endpoint != None:
        first = True
        for temp_node in temp_list:
            absolute_dist = abs(temp_node.loc[0] - endpoint[0]) + abs(
                temp_node.loc[1] - endpoint[1])
            if first:
                final_node = temp_node
                final_dist = absolute_dist
                first = False
            elif absolute_dist < final_dist:
                final_node = temp_node
                final_dist = absolute_dist
            else:
                continue    
        return(final_node)
    
    # Random tie break if random set to True.
    elif len(temp_list) > 1 and random == True:
        final_node = np.random.choice(np.array(temp_list))
        return(final_node)
    # No tie break required.
    elif len(temp_list) == 1:
        return(temp_list[0])


def greedy(startnode:Node, terrain, endpoint = None, return_visited = False, random = False):
    '''
    startnode:node -> The node at which pathfinding begins from.
    terrain -> A numpy array representing the map through which pathfinding
               will occur.
    endpoint -> A co-ordiante (x,y) on the graph to which the most optimum path
                will be found for.
    Function returns; (A list of co-ordianates representing the path, distance
                      of the path)

    '''
    shape = terrain.shape
    visited = np.zeros((shape), dtype = int) 
    pathlist = [[startnode]]
    start = True
    finished = False
    while finished == False:
        
        if start == True:
            queue = [startnode]
            visited[startnode.loc] = 1
            start = False
        # A queue system to determine which node's neighbours get checked next.
        if random == False:
            temp = neighbours_greedy(queue[0], terrain, visited,
                                        endpoint = endpoint)
        elif random == True:
            temp = neighbours_greedy(queue[0], terrain, visited, True,
                                        endpoint = endpoint)
        if queue[0].loc == endpoint:
            break
        if temp != None:
            visited[temp.loc] = 1
            queue.append(temp)
            
            # This loop creates a list of every path taken.
            # Not best method for greedy, carried over from dijkstra function.
            temp = [temp]
            for path in pathlist:
                for i in range(0, len(temp)):
                    x = path[:]
                    if queue[0] == x[-1]:
                        x.append(temp[i])
                        pathlist.append(x)
            queue.pop(0)
    
            # A patchwork solution of removing nodes from the queue, if a node
            #with the same location but a smaller distance exists in the queue.
            for node in queue:
                for node2 in queue:
                    if node.loc == node2.loc and node.dist < node2.dist:
                        queue.remove(node2)
                    else:
                        continue
        else:
            queue.pop(0)
        # If there are no more possible paths, terminate.
        
        if len(queue) == 0:
            finished = True
            
    pre_final_path = pathlist[-1]
    # Reformates paths into pure tuple co-ordiantes and also returns distance
    # seperately.
    final_path = []
    for node in pre_final_path:
        final_path.append(node.loc)        
    if len(pre_final_path) != 0:
        final_dist = pre_final_path[-1].dist
    else:
        final_dist = 'Null'
    
    if return_visited == False:
        return(final_path, final_dist)
    if return_visited == True:
        unique, counts = np.unique(visited, return_counts=True)
        x = dict(zip(unique, counts))
        total = x[1]
        return(final_path, final_dist, total)
    
```

```python
# =============================================================================
# This is the A* algorithm. It may not be the best code as it is a modified
# version of my dijkstra algorithm.
# =============================================================================
def neighbours_a_star(n:StarNode, terrain, visited, weight, endpoint:tuple = None,
                        check:dict = None): 
    '''
    Finds all valid neighbour StarNodes next to target StarNode(n).
    Takes in a map (terrain) and corresponding bool map (visited).
    Only works within the search function.
    Returns a list of StarNode classes that are neighbour StarNodes
    to the input StarNode.
    '''
    # Create a new StarNode for each direction
    north = StarNode(n.loc[0] - 1, n.loc[1], n.dist)
    south = StarNode(n.loc[0] + 1, n.loc[1], n.dist)
    east = StarNode(n.loc[0], n.loc[1] + 1, n.dist)
    west = StarNode(n.loc[0], n.loc[1] - 1, n.dist)
    neighbour_list = []
    shape = terrain.shape
    length, width = shape[0], shape[1]
    for i in [north, south, west, east]:
        direction = (i.loc[0], i.loc[1])
        r,c = direction[0], direction[1]
        
        absolute_dist = (abs(i.loc[0] - endpoint[0]) + abs(i.loc[1] - endpoint[1])) * weight
        i.new_est(absolute_dist)
        
        # Checking that the new StarNodes are not out of bounds
        if r != -1 and c != -1 and r != length and c != width:
            # Checking that the new Nodes are not walls or previously visited.
            if terrain[direction] != 0 and visited[direction] == 0:
                temp = StarNode(i.loc[0], i.loc[1],
                            (i.dist + terrain[i.loc]))
                neighbour_list.append(temp)
            # This allows for Nodes to be overriden at a location, if they
            # have a smaller distance than the existing Node at that location.
            elif i.loc in check:
                x = check[i.loc]
                if (int(i.total) + terrain[i.loc]) < int(x.total):
                    if terrain[direction] != 0 and visited[direction] == 1:
                        temp = StarNode(i.loc[0], i.loc[1],
                                    (i.dist + terrain[i.loc]))
                        neighbour_list.append(temp)
    
    for node in neighbour_list:
        absolute_dist = (abs(node.loc[0] - endpoint[0]) + abs(node.loc[1] - endpoint[1])) * weight
        node.new_est(absolute_dist)
        
    return(neighbour_list)


def a_star(startnode:StarNode, terrain, endpoint = None, return_visited = False,
           weight = 1, beam = 0):
    '''
    startnode:StarNode -> The StarNode at which pathfinding begins from.
    terrain -> A numpy array representing the map through which pathfinding
               will occur.
    endpoint -> A co-ordinate (x,y) on the graph to which the most optimum path
                will be found for.
    return_visited -> will return count of how many nodes visited.
    weight -> Default 1 (normal A*). Changes the weight of the heuristic by
              multiplication. Values closer to 0 are more like dijkstra,
              values closer to infinity are more like greedy first.
    beam -> Default 0 (No beam). Implements the beam algorithm which works by
            limiting the amount of nodes allowed in the queue.

    Function returns; tuple(A list of co-ordinates representing the path,
                            distance of the path)

    '''
    shape = terrain.shape
    visited = np.zeros((shape), dtype = int) 
    pathlist = [[startnode]]
    nodedict = {startnode.loc:startnode}
    start = True
    finished = False

    if endpoint == None:
        print('Need an endnode')
        return()

    while finished == False:
        if start == True:
            queue = [startnode]
            visited[startnode.loc] = 1
            start = False
        # Break out of loop when reaching destination
        if queue[0].loc == endpoint:
            finished = True
        # A queue system to determine which node's neighbours get checked next.
        temp = neighbours_a_star(queue[0], terrain, visited, weight, endpoint, 
                                 nodedict)
        for new_node in temp:
            visited[new_node.loc] = 1
            queue.append(new_node)
            if new_node in nodedict.keys():
                nodedict.pop(new_node.loc)
            nodedict.update({new_node.loc:new_node})
        #This loop creates a list of every path taken.
        for path in pathlist:
            for i in range(0, len(temp)):
                x = path[:]
                if queue[0] == x[-1]:
                    x.append(temp[i])
                    pathlist.append(x)
        queue.pop(0)
        queue.sort()

        # A patchwork solution of removing nodes from the queue, if a node
        # with the same location but a smaller distance exists in the queue.
        for node in queue:
            for node2 in queue:
                if node.loc == node2.loc and node.total < node2.total:
                    queue.remove(node2)
                else:
                    continue

        # Part of the beam algorithm. Limits size of queue by removing nodes.
        if beam > 0 and beam < len(queue):
            queue = queue[0:beam]
        # If there are no more possible paths, terminate.
        if len(queue) == 0:
            finished = True

    # This finds the path that has the endpoint and has the lowest distance.
    final_list = []
    first = True
    for path in pathlist:
        if path[-1].loc == endpoint:
            if first == True:
                final_list = path[:]
            elif path[-1].dist < final_list[-1].dist:    
                final_list = path[:]
    
    # Reformats paths into pure tuple co-ordinates and also returns distance
    # separately.
    final_path = []
    for StarNode in final_list:
        final_path.append(StarNode.loc)        
    if len(final_list) != 0:
        final_dist = final_list[-1].dist
    else:
        final_dist = 'Null'
    if return_visited == False:
        return(final_path, final_dist)
    if return_visited == True:
        unique, counts = np.unique(visited, return_counts=True)
        x = dict(zip(unique, counts))
        total = x[1]
        return(final_path, final_dist, total)
```
